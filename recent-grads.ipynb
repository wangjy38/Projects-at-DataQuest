```python
Lowest_ShareWoman_majors=recent_grads.sort_values("ShareWomen").head(10)["Major"]
Highest_ShareWoman_majors=recent_grads.sort_values("ShareWomen", ascending=False).head(10)["Major"]
```

```python
ax11=recent_grads[recent_grads["Major"].isin(Lowest_ShareWoman_majors)].plot.bar(x="Major", y="ShareWomen")
ax11.set_title("ShareWomen in 10 lowest paying majors")
ax11.set_ylim(0,1)
```

```python
ax12=recent_grads[recent_grads["Major"].isin(Highest_ShareWoman_majors)].plot.bar(x="Major", y="ShareWomen")
ax12.set_title("ShareWomen in 10 highest paying majors")
```

```python
ax13=recent_grads[recent_grads["Major"].isin(Lowest_ShareWoman_majors)].plot.bar(x="Major", y="Unemployment_rate")
ax13.set_title("Unemployment_rate in 10 lowest paying majors")
ax13.set_ylim(0,0.15)
```

```python
ax14=recent_grads[recent_grads["Major"].isin(Highest_ShareWoman_majors)].plot.bar(x="Major", y="Unemployment_rate")
ax14.set_title("Unemployment_rate in 10 highest paying majors")
ax14.set_ylim(0,0.15)
```
